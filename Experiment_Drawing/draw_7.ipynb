# 优化后的电路的情况

- 深度降低了，CNOT门数量减少了，所需的比特数降低了

```python
import pandas as pd
import matplotlib.pyplot as plt

#input_file = f"result/circuit/ciruit_results_mean.csv"
input_file = f"result/circuit/ciruit_results_mean_10000.csv"
# 读取 CSV 文件
df = pd.read_csv(input_file)

# 提取数据
num_tickers = df['num_tickers']
depth_decrease = df['depth_decrease']  # decrease: 优化后的值 / 原先的值
CNOT_decrease = df['CNOT_decrease']
qubit_num_decrease = df['qubit_num_decrease']

# 创建画布和子图
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(8, 12))

# 绘制 depth_decrease 折线图
axes[0].plot(num_tickers, depth_decrease, label='Depth Decrease')
axes[0].set_ylabel('Depth Decrease')
axes[0].legend()
axes[0].grid(True)

# 绘制 CNOT_decrease 折线图
axes[1].plot(num_tickers, CNOT_decrease, label='CNOT Decrease')
axes[1].set_ylabel('CNOT Decrease')
axes[1].legend()
axes[1].grid(True)

# 绘制 qubit_num_decrease 折线图
axes[2].plot(num_tickers, qubit_num_decrease, label='Qubit Num Decrease')
axes[2].set_xlabel('Number of Stocks')
axes[2].set_ylabel('Qubit Num Decrease')
axes[2].legend()
axes[2].grid(True)

# 设置标题
fig.suptitle('Decrease Values vs. num_tickers', fontsize=16)

# 调整子图之间的间距
plt.tight_layout()

# 设置图像的分辨率
dpi = 300  # 设置dpi值
# 保存图像
output_filename = f"image/circuit/result_v1_all_in_one.pdf"
plt.savefig(output_filename, dpi=dpi, bbox_inches='tight')

# 显示图形
plt.show()
```
