```python
import pandas as pd

# 读取 CSV 文件
input_file = f"./ciruit_results.csv"
df = pd.read_csv(input_file)

# 计算相对减少比率
df['depth_decrease'] = df['optimized_depth'] / df['original_depth']
df['CNOT_decrease'] = df['optimized_gate_count'] / df['original_gate_count']
df['qubit_num_decrease'] = df['optimized_num_qubit'] / df['original_num_qubit']

# 按 num_tickers 聚合，并计算 optimized_depth 相对减少比率的平均值
result_df = df.groupby('num_tickers')[['depth_decrease', 'CNOT_decrease', 'qubit_num_decrease']].mean().reset_index()

# 将结果保存到另一个 CSV 文件
output_file = f"./ciruit_results_mean.csv"
result_df.to_csv(output_file, index=False)
print(f"finished in {output_file}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

input_file = f"./ciruit_results_mean.csv"
# 读取 CSV 文件
df = pd.read_csv(input_file)

# 提取数据
num_tickers = df['num_tickers']
depth_decrease = df['depth_decrease']
CNOT_decrease = df['CNOT_decrease']
qubit_num_decrease = df['qubit_num_decrease']

# 创建画布和子图
fig, ax = plt.subplots()

# 绘制三个折线图
ax.plot(num_tickers, depth_decrease, label='Depth Decrease')
ax.plot(num_tickers, CNOT_decrease, label='CNOT Decrease')
ax.plot(num_tickers, qubit_num_decrease, label='Qubit Num Decrease')

# 设置图例
ax.legend()

# 设置横纵轴标签和标题
ax.set_xlabel('num_tickers')
ax.set_ylabel('Decrease Value')
ax.set_title('Decrease Values vs. num_tickers')

# 显示网格线
ax.grid(True)

# 显示图形
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

input_file = f"./ciruit_results_mean.csv"
# 读取 CSV 文件
df = pd.read_csv(input_file)

# 提取数据
num_tickers = df['num_tickers']
depth_decrease = df['depth_decrease']
CNOT_decrease = df['CNOT_decrease']
qubit_num_decrease = df['qubit_num_decrease']

# 创建画布和子图
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(8, 12))

# 绘制 depth_decrease 折线图
axes[0].plot(num_tickers, depth_decrease, label='Depth Decrease')
axes[0].set_ylabel('Depth Decrease')
axes[0].legend()
axes[0].grid(True)

# 绘制 CNOT_decrease 折线图
axes[1].plot(num_tickers, CNOT_decrease, label='CNOT Decrease')
axes[1].set_ylabel('CNOT Decrease')
axes[1].legend()
axes[1].grid(True)

# 绘制 qubit_num_decrease 折线图
axes[2].plot(num_tickers, qubit_num_decrease, label='Qubit Num Decrease')
axes[2].set_xlabel('num_tickers')
axes[2].set_ylabel('Qubit Num Decrease')
axes[2].legend()
axes[2].grid(True)

# 设置标题
fig.suptitle('Decrease Values vs. num_tickers', fontsize=16)

# 调整子图之间的间距
plt.tight_layout()

# 显示图形
plt.show()
```
